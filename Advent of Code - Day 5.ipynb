### Part 1

```python
import pandas as pd

#pull in data from txt (keeping blank rows)
with open('Day5_Part1_input.txt') as f:
    lines = f.readlines()
df_raw=pd.read_csv('Day5_Part2_input.txt',sep="  ", header=None, skip_blank_lines=False)

#pull in data from txt (keeping blank rows)
full_s=''
j=1
while j < len(lines[0]):
    i=0
    while i < len(lines):
        full_s+=lines[i][j]
        i+=1
    full_s+=" "
    j+=4
full_s.split()
```

```python
df_steps=pd.read_csv('Day5_Part2_input.txt',sep="  ", header=None, skip_blank_lines=False)
df_steps['move'] = df_steps[0].str[5:7].astype('int')
df_steps['from'] = df_steps[0].str[-6].astype('int')
df_steps['to'] = df_steps[0].str[-1].astype('int')
df_steps
l_move=df_steps['move'].to_list()
l_from=df_steps['from'].to_list()
l_to=df_steps['to'].to_list()
df_steps.head(20)
```

```python
str_array=full_s.split()
curr_step=0
n_steps = len(df_steps['to'])
while curr_step < n_steps:
    count_moves=0
    n_moves , n_from, n_to = l_move[curr_step], l_from[curr_step], l_to[curr_step]
    n_from=n_from-1
    n_to=n_to-1
    while count_moves < n_moves:

        str_array[n_to]=''.join((str_array[n_from][0],str_array[n_to]))
        str_array[n_from]=str_array[n_from][1:]       
        count_moves+=1
    curr_step+=1

str_final=''
for ind_str in str_array:
    str_final+=ind_str[0]
print(str_final)
```

### Part 2

```python
str_array=full_s.split()

curr_step=0
n_steps = len(df_steps['to'])
while curr_step < n_steps:
    count_moves=0
    n_moves , n_from, n_to = l_move[curr_step], l_from[curr_step], l_to[curr_step]
    n_from=n_from-1
    n_to=n_to-1
    n_moves=n_moves
    
    str_array[n_to]=''.join((str_array[n_from][:n_moves],str_array[n_to]))
    str_array[n_from]=str_array[n_from][n_moves:]       
    curr_step+=1

str_final=''
for ind_str in str_array:
    str_final+=ind_str[0]
print(str_final)
```
